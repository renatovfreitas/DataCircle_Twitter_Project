```python
import re
import pandas as pd

# Function to remove URLs from text with handling for non-string values
def remove_urls(text):
    # Check if the input is a string
    if isinstance(text, str):  
        url_pattern = re.compile(r'http[s]?://\S+')
        # Remove URLs from the text
        return url_pattern.sub('', text)
    else:
        # If the input is not a string (e.g., NaN or float), return it as is
        return text

# Assume 'df' is your DataFrame and it has a 'tweet' column

# Apply the remove_urls function to each tweet in the 'tweet' column
df['tweet_cleaned'] = df['tweet'].apply(remove_urls)

# Optionally, print the original and cleaned tweets to see the result
print(df[['tweet', 'tweet_cleaned']])
```
