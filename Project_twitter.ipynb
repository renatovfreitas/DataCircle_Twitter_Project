Ensure Data Consistency

    Standardizing date/time formats
    Ensuring consistent text formatting (e.g., lowercasing)
    Cleaning up usernames and hashtags (removing extra spaces or symbols)

# Standardize Date/Time (created_at)
This ensures that all dates are in a consistent format.


```python

df_combined['created_at'] = pd.to_datetime(df_combined['created_at'], errors='coerce')
```

# Convert 'likes', 'user_followers_count', 'lat', 'long' to numeric types

```python
# Convert 'likes', 'user_followers_count', 'lat', 'long' to numeric types
df_combined['likes'] = pd.to_numeric(df_combined['likes'], errors='coerce')
df_combined['user_followers_count'] = pd.to_numeric(df_combined['user_followers_count'], errors='coerce')
df_combined['lat'] = pd.to_numeric(df_combined['lat'], errors='coerce')
df_combined['long'] = pd.to_numeric(df_combined['long'], errors='coerce')
```

```python
# Check for any null values after conversion
print(df_combined[['likes', 'user_followers_count', 'lat', 'long']].isnull().sum())
```

# Standardize categorical columns
Text columns like tweet, user_name, user_screen_name, etc., are converted to lowercase and trimmed of any extra spaces.

```python
# Standardize categorical columns
df_combined['source'] = df_combined['source'].str.lower().str.strip()
df_combined['city'] = df_combined['city'].str.lower().str.strip()
df_combined['country'] = df_combined['country'].str.lower().str.strip()
df_combined['state'] = df_combined['state'].str.lower().str.strip()
df_combined['state_code'] = df_combined['state_code'].str.upper().str.strip()
```

```python
text_columns = ['tweet', 'user_name', 'user_screen_name', 'user_description', 'user_location', 'source']
for col in text_columns:
    df_combined[col] = df_combined[col].str.lower().str.strip()
```

```python
# Check for any null values after conversion
print(df_combined[text_columns].isnull().sum())
```

# Top 5 Countries with Most Tweets for Each Candidate

```python
# Group by country and candidate to count the tweets
country_tweet_counts = df_combined.groupby(['country', 'candidate']).size().reset_index(name='tweet_count')

# Sort values to get top countries
sorted_counts = country_tweet_counts.sort_values(by=['candidate', 'tweet_count'], ascending=[True, False])
```

```python
# Get the top 5 countries for each candidate
top_5_countries_per_candidate = sorted_counts.groupby('candidate').head(5).reset_index(drop=True)
```

```python
# Display the top 5 countries
print(top_5_countries_per_candidate)
```

```python
# Plotting
plt.figure(figsize=(10, 5))
sns.barplot(x='tweet_count', y='country', hue='candidate', data=top_5_countries_per_candidate, palette='Set2')
plt.title('Top 5 Countries with Most Tweets for Each Candidate')
plt.xlabel('Number of Tweets')
plt.ylabel('Country')
plt.legend(title='Candidate')
plt.show()
```

# What Time Do Users Tweet the Most?

```python

df_combined['tweet_hour'] = df_combined['created_at'].dt.hour
# Count the number of tweets for each hour and candidate
hourly_tweet_counts = df_combined.groupby(['tweet_hour', 'candidate']).size().reset_index(name='tweet_count')
```

```python
# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_tweet_counts, x='tweet_hour', y='tweet_count', hue='candidate', marker='o')
plt.title('Number of Tweets by Hour (24-hour Format)')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Tweets')
plt.xticks(range(0, 24))  # Show all hours in the x-axis
plt.grid(True)
plt.legend(title='Candidate')
plt.show()
```

# cities in the U.S. with the most tweets for each candidate (Biden and Trump)

```python
# Fill NaN values in the 'country' column with an empty string
df_combined['country'] = df_combined['country'].fillna('')

# Filter the DataFrame for tweets from the United States
df_us = df_combined[df_combined['country'].str.lower().str.contains('united states')]
```

```python
# Count the number of tweets for each city and candidate
city_tweet_counts = df_us.groupby(['city', 'candidate']).size().reset_index(name='tweet_count')

# Get the top 10 cities for each candidate
top_cities_per_candidate = (
    city_tweet_counts
    .sort_values(['candidate', 'tweet_count'], ascending=[True, False])
    .groupby('candidate')
    .head(10)  # Take top 10 for each candidate
)

# Reset the index for better plotting
top_cities_per_candidate.reset_index(drop=True, inplace=True)
```

```python
# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='tweet_count', y='city', hue='candidate', data=top_cities_per_candidate, palette='Set2')
plt.title('Top 10 Cities with Most Tweets for Each Candidate (U.S. Only)')
plt.xlabel('Number of Tweets')
plt.ylabel('City')
plt.legend(title='Candidate')
plt.show()
```
